```python
#자판기 메뉴 프로그램

def sell_menu() :
  print('** 자판기 판매 메뉴 **')
  print()
  for i in range(5) :
    print(i+1, '{} {}'.format(menu[i], price[i]))

def drink(money) :
  while True :
    num = int(input('메뉴 번호 선택 : '))
    if num == 0 : break

    if price[num-1] > money :
      print('잔액부족')

    else :
      print('{} 구입 완료'.format(menu[num-1]))
      print('잔액 : {}'.format(money - price[num-1]))
      money = money - price[num-1]
  return money

def change(money) :
  print('자판기 종료')
  coin500 = money / 500
  money = money % 500
  coin100 = money / 100
  print('coin500 : {}\ncoin100 : {}'.format(coin500, coin100))



# 메인코드
menu = ['콜라', '사이다', '환타', '커피', '생수']
price = [500, 500, 600, 600, 400]

sell_menu()

money = int(input('돈을 투입하세요 : '))
print()

money = drink(money)
change(money)
```

```python
# 출결 프로그램

def save(day) :
  result = []
  for i in range(day) :
    print('%d주차 수업에 참여하셨나요? ' %(i+1))
    result.append(int(input('출석은 1, 결석은 0 >>>')))
  return result

def count(result) :
 cnt1, cnt2  =  result.count(1), result.count(0)
 return cnt1, cnt2

def percent(cnt1, cnt2, day) :
  cnt1p, cnt2p =  cnt1 / day * 100, cnt2 / day * 100
  return cnt1p, cnt2p

def print_result(cnt1,cnt2,cnt1p,cnt2p) :
  print('출석 : %d, 결석 : %d' % (cnt1, cnt2))
  print('출석률 : %.2f%%, 결석률 : %.2f%%' % (cnt1p, cnt2p))
  if cnt2p > 30:
    print('경고! 수업일수 부족')


day = int(input('한학기 수업 주차 입력 : '))
result = save(day)
cnt1, cnt2 = count(result)
cnt1p, cnt2p = percent(cnt1, cnt2, day)
print_result(cnt1,cnt2,cnt1p,cnt2p)
```

```python
import random

def play_rps(user_choice, com_choice):
    if user_choice == com_choice:
        return (0, 0)  # 비김

    # 승패 판단
    if user_choice == '가위':
        if com_choice == '바위':
            return (0, 1)  # 사용자 패배
        else:
            return (1, 0)  # 사용자 승리
    elif user_choice == '바위':
        if com_choice == '보':
            return (0, 1)
        else:
            return (1, 0)
    elif user_choice == '보':
        if com_choice == '가위':
            return (0, 1)
        else:
            return (1, 0)
    else:
        return None  # 잘못된 입력

# 게임 시작
options = ['가위', '바위', '보']
user_win = 0
com_win = 0

print('가위 바위 보 게임 시작 (3승 먼저 달성하면 종료)')

while user_win < 3 and com_win < 3:
    user = input('가위 바위 보 중 하나를 선택하세요 (종료 0) : ')
    if user == '0':
        print('게임 종료')
        break

    com = random.choice(options)
    result = play_rps(user, com)

    # 승패 업데이트
    user_win += result[0]
    com_win += result[1]

    # 결과 출력
    if result == (1, 0):
        print(f'컴퓨터: {com}, 사용자: {user} → 이겼습니다!')
    elif result == (0, 1):
        print(f'컴퓨터: {com}, 사용자: {user} → 졌습니다!')
    else:
        print(f'컴퓨터: {com}, 사용자: {user} → 비겼습니다!')

    print(f'현재 스코어 → 사용자 {user_win} : 컴퓨터 {com_win}\n')

# 최종 승리 출력
if user_win == 3:
    print(f'사용자가 3승 달성! 게임 종료')
elif com_win == 3:
    print(f'컴퓨터가 3승 달성! 게임 종료')
```

```python
# 평점평균 계산 시스템

def graduate(gpa) :
  tot_smemster = int(input('\n총 등록 학기수 입력 : '))
  if tot_smemster >= 8 :
    print('졸업학기 충족')
  else :
    print('졸업학기 부족')

  tot_credit = int(input('\n수강 완료 학점수 입력 : '))
  if tot_credit >= 120 :
    print('졸업학점 충족')
  else :
    print('졸업학점 부족')

  print('\n평균 평점 : %.2f' %gpa)
  if gpa >= 2.5 :
    print('평균평점 충족')
  else :
    print('평균평점 부족')
  if tot_smemster >= 8 and tot_credit >= 120 and gpa >= 4.0 :
    print('졸업 장학 대상자!!')

def subject() :
  while True :
    title = input('과목명(0:종료)')
    if title == '0' :
      break
    credit = int(input('학점 수 : '))
    grade = input('취득학점(A,B,C,D)')
    course.append([title,credit,grade])

def avg_grade() :
  total = 0
  result = 0
  for i in course :
    if i[2] == 'A' :
      total += i[1]
      result += i[1] * 4.5
    elif i[2] == 'B' :
      total += i[1]
      result += i[1] * 3.5
    elif i[2] == 'C' :
      total += i[1]
      result += i[1] * 2.5
    elif i[2] == 'F' :
      total += i[1]
  return result / total


course = []
while True :
  choice = eval(input('1.수강 강좌 정보 입력 2. 평점평균 확인 3. 졸업 가능 여부 확인 0.종료'))
  if choice < 0 or choice > 3 :
    print('1~3의 숫자를 입력하세요.')
    continue

  elif choice == 1 :
    print('<수강 강좌 정보 입력>')
    subject()
    print('<수강 강좌 정보 입력 종료')

  elif choice == 2 :
    print('<수강 강좌 목록>')
    print('과목명  학점수  학점')
    for i in course :
      print(f"{i[0]}  {i[1]}  {i[2]}")
    gpa = avg_grade()
    print('평점 평균 : %.2f' %gpa)
    if gpa >= 4.0 :
      print('장학 대상자!!')

  elif choice == 3 :
    print('<졸업 가능 여부 확인>')
    graduate(gpa)
    print('<졸업 가능 여부 확인 종료>')
    print()

  elif choice == 0 :
    print('평점 평균 계산 시스템 종료')
    break

```

```python
# 연락처 관리 프로그램
dic = {}
while True :
  choice = int(input('\n1. 연락처 추가\n2. 연락처 삭제\n3. 연락처 수정\n4. 연락처 조회\n5. 전체 연락처 출력 \n0. 종료\n'))

  if choice == 1 :
    name = input('이름을 입력하세요.')
    phone = input('전화번호를 입력하세요.')
    if name in dic :
      print('이미 존재하는 이름이빈다.')
    else :
      dic[name] = phone
      print('연락처가 추가되었습니다.')


  elif choice == 2 :
    rm = input('삭제할 연락처의 이름을 입력해주세요. ')
    if rm in dic :
      del dic[rm]
      print('연락처가 삭제 되었습니다.')
    elif rm not in dic :
      print('연락처가 존재하지 않습니다.')

  elif choice == 3 :
    name2 = input('수정할 연락처의 이름을 입력해주세요')
    if name2 in dic :
      dic[name2] = input('수정할 전화번호를 입력해주세요.')
      print('연락처가 수정되었습니다.')
    elif name2 not in dic :
      print('연락처가 존재하지 않습니다.')

  elif choice == 4 :
    search = input('조회할 연락처의 이름을 입력해주세요. ')
    if search in dic :
      print(dic[search])
    elif search not in dic :
      print('연락처가 존재하지 않습니다.')

  elif choice == 5 :
    print(dic)

  elif choice == 0 :
    print('연락처 관리 프로그램 종료')
    break


```
