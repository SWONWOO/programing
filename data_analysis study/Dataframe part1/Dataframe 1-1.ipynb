```python
# 판다스 데이터프레임

from pandas import DataFrame

# 딕셔너리 생성
data = {
    '종목코드' : ['037730', '036360', '005760'],
    '종목명' : ['3R', '3SOFT', 'ACTS'],
    '현재가' : [1510, 1790, 1185]
}

df = DataFrame(data)
print(df)
print()

# 리스트 생성
data = [
    ['037730', '3R', 1510],
    ['036360', '3SOFT', 1790],
    ['005760', 'ACTS', 1185]
]

columns = ['종목코드', '종목명', '현재가']

df = DataFrame(data = data, columns = columns)
print(df)
print()

```

```python
# 데이터프레임 인덱스

from pandas import DataFrame

data = {
    '종목코드' : ['037730', '036360', '005760'],
    '종목명' : ['3R', '3SOFT', 'ACTS'],
    '현재가' : [1510, 1790, 1185],
    '등락률' : [7.36, 1.65, 1.28]
}

df = DataFrame(data)
print(df)
# 종목코드를 인덱스로 지정
df = df.set_index('종목코드')
print(df)
print()

# 데이터프레임 생성할때 인덱스 지정

data = [
    ['3R', 1510, 7.36],
    ['3SOFT', 1790, 1.65],
    ['ACTS', 1185, 1.28]
]

index = ['037730', '036360', '005760']


columns = ['종목명', '현재가', '등락률']

df = DataFrame(data = data, index = index, columns = columns)
df.index.name = '종목코드'
print(df)
```

```python
# 컬럼 인덱싱

from pandas import DataFrame

data = [
    ['3R', 1510, 7.36],
    ['3SOFT', 1790, 1.65],
    ['ACTS', 1185, 1.28]
]

index = ['037730', '036360', '005760']
columns = ['종목명', '현재가', '등락률']
df = DataFrame(data = data, index = index, columns = columns)
print(df['현재가'])
print(df.현재가) # 위와 동일 컬럼의 이름만으로 컬럼 시리즈 선택 가능
data1 = ['현재가', '등락률']
print(df[data1])
print()

# 인덱싱 결과 시리즈, 데이터프레임
print(df['현재가']) # 결과 = 시리즈
print(df[['현재가']]) # 결과 = 데이터프레임
```

```python
# 로우 인덱싱

from pandas import DataFrame

data = [
    ['3R', 1510, 7.36],
    ['3SOFT', 1790, 1.65],
    ['ACTS', 1185, 1.28]
]

index = ['037730', '036360', '005760']
columns = ['종목명', '현재가', '등락률']
df = DataFrame(data = data, index = index, columns = columns)
print(df.loc['037730']) # loc 속성 (인덱스를 기준으로 로우 데이터 추출)
print(df.iloc[0])
print(df.iloc[-1]) # iloc 속성 (행번호를 기준으로 로우 데이터 추출)
print()

print(df.loc[['037730', '036360']])
print(df.iloc[[0, -1]])
```

```python
# 특정 값 가져오기

from pandas import DataFrame

data = [
    ['3R', 1510, 7.36],
    ['3SOFT', 1790, 1.65],
    ['ACTS', 1185, 1.28]
]

index = ['037730', '036360', '005760']
columns = ['종목명', '현재가', '등락률']
df = DataFrame(data = data, index = index, columns = columns)

# 행번호로 행 선택후 시리즈 인덱싱
print(df.iloc[0].iloc[1])
print(df.iloc[0].loc['현재가'])
print(df.iloc[0]['현재가'])

# 인덱스로 행 선택후 시리즈 인덱싱
print(df.loc['037730'].iloc[1])
print(df.loc['037730'].loc['현재가'])
print(df.loc['037730']['현재가'])
print()

# 한번에 입력하기
print(df.loc['037730', '현재가'])
print(df.iloc[0,1])
print()

# 컬럼 인덱싱 하고 시리즈 인덱싱
print(df['현재가'].iloc[0])
print(df.현재가.loc['037730'])
print(df.현재가['037730'])

# 가독성이나 실행속도 측면에서 df.loc[행,열] 또는 df.iloc[행번호, 열번호] 방식이 더 좋다
```

```python
# 특정 범위 가져오기

from pandas import DataFrame

data = [
    ['3R', 1510, 7.36],
    ['3SOFT', 1790, 1.65],
    ['ACTS', 1185, 1.28]
]

index = ['037730', '036360', '005760']
columns = ['종목명', '현재가', '등락률']
df = DataFrame(data = data, index = index, columns = columns)

print(df.loc[['037730','036360']])
print(df.iloc[[0,1]])
print()

df = df.iloc[[0,1]]
df = df[['종목명','현재가']]
print(df)
```
