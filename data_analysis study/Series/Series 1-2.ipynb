```python
# 시리즈와 Map

from pandas import Series

s = Series(['1,234', '2,345','3,456'])

def remove_comma(x) :
  return int(x.replace(',', ''))

result = s.map(remove_comma)
print(result)

def is_greater_than_2000(x) :
  if x > 2000 :
    return '크다'
  else :
    return '작다'

result = result.map(is_greater_than_2000)
print(result)
```

```python
# 시리즈 필터링

from pandas import Series

data = [42500, 42550, 41800, 42550, 42650]
index = ['2019-05-31', '2019-05-30', '2019-05-29', '2019-05-28', '2019-05-27']
s = Series(data = data, index = index)
cond = s > 42000
print(cond)
print(s[cond])
print()

close = [42500, 42550, 41800, 42550, 42650]
open = [42600, 42200, 41850, 42550, 42500]
index = ['2019-05-31', '2019-05-30', '2019-05-29', '2019-05-28', '2019-05-27']
oepn = Series(data = open, index = index)
close = Series(data = close, index = index)

cond = close > open
print(cond)
print(close[cond]) # = close[close > open]
print(close[cond].index)
print()

diff = close - open
print(diff)
print(diff[cond])
```

```python
# 시리즈 정렬 및 순위

from pandas import Series

# sort_values() 정렬
data = [3.1, 2.0, 10.1, 5.1]
index = ['000010', '000020', '000030', '000040']
s = Series(data = data, index = index)

print(s.sort_values()) # 오름차순 정렬
print(s.sort_values(ascending = False)) # 내림차순 정렬
print()

# rank() 순위

print(s.rank()) # 데이터 값이 작은 순위
print(s.rank(ascending = False)) # 데이터 값이 큰 순위
```
