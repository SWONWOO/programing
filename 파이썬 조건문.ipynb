```python
# if ~ else문

member = input('회원이십니까? (Y/N) : ')

if member == 'Y' :
  print('어서오세요!')

else :
  print('회원가입을 하고 오세요!')
```

```python
# if ~ elif ~ else문

age = int(input('나이를 입력하세요 : '))
price = 20000

if 1<= age < 6 :
  print('입장료는 무료')

elif 6 <= age < 60 :
  print('입장료는', price, '원')

elif age >= 60 :
  print('입장료는', price * 0.5,'원')

else :
  print('1살 이상의 나이를 입력해주세요.')

```

```python
# if ~ in ~ elif ~ else문

pocket = ['스마트폰', '교통카드', '현금']
item = input('결제 수단을 입력해주세요 : ')

if item in pocket and item == '현금' :
  print('택시를 이용하세요.')

elif item in pocket and item =='교통카드' :
  print('버스를 이용하세요.')

else :
  print('걸어가세요.')
```

```python
# 중첩 if 문 (가위바위보 예제)

import random
list = ['가위', '바위', '보']
com = random.choice(list)
player = input('가위 바위 보 ! ')

if player == '가위' :
  if com == '가위' :
    result = '비김'
  elif com == '바위' :
    result = '졌음'
  elif com == '보' :
    result = '이김'

elif player == '바위' :
  if com == '가위' :
    result = '이김'
  elif com == '바위' :
    result = '비김'
  else :
    result = '졌음'

else :
  if com == '가위' :
    result = '졌음'
  elif com == '바위' :
    result = '이김'
  else :
    result = '비김'


if player in list :
  print('컴퓨터 : %s, 당신 : %s \n 결과 : %s' %(com, player, result))

else :
  print('가위, 바위, 보중 한개를 입력하세요.')
```

```python
#사칙연산 결과 출력 프로그램

a = int(input('첫 번째 수 입력 : '))
b = int(input('두 번째 수 입력 : '))
c = input('사칙연산 중 하나를 입력하시오. (+, -, *, /)')

if c == '+' :
  print('%d + %d = %d' %(a,b,a + b))
elif c == '-' :
  print('%d - %d = %d' %(a,b,a - b))
elif c == '*' :
  print('%d * %d = %d' %(a,b,a * b))
elif c == '/' :
  print('%d / %d = %d' %(a,b,a / b))
else :
  print('사칙연산중 하나를 입력해주세요.')


```

```python
# 합격 불합격 여부 출력 프로그램

sc_1 = int(input('1차 점수 입력 : '))
sc_2 = int(input('2차 점수 입력 : '))

if sc_1 >= 50 and sc_2 >= 50 :
  if (sc_1 + sc_2) / 2 >= 70 :
    print('합격')
  else :
    print ('불합격')

else :
  print('불합격')
```

```python
# 학점 출력 프로그램

score = int(input('점수 입력 :'))

if score >= 90 :
  result = 'A학점'
elif 80 <= score < 90 :
  result = 'B학점'
elif 70 <= score < 80 :
  result = 'C학점'
elif 60 <= score < 70 :
  result = 'D학점'
else :
  result = 'F학점'

print('%d점은 %s입니다.' %(score,result))
```
